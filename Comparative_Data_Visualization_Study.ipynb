```python
import pandas as pd
import matplotlib.pyplot as plt

df1 = pd.DataFrame({'Category': ['A', 'B', 'C'], 'Values': [10, 20, 30]})
df2 = pd.DataFrame({'Category': ['A', 'B', 'C'], 'Values': [15, 25, 20]})

x = range(len(df1['Category']))
width = 0.35

plt.figure(figsize=(10, 5))
plt.bar(x, df1['Values'], width, label='Dataset 1', color='blue')
plt.bar([i + width for i in x], df2['Values'], width, label='Dataset 2', color='orange')

plt.xlabel('Category')
plt.ylabel('Values')
plt.title('Comparative Data Visualization')
plt.xticks([i + width/2 for i in x], df1['Category'])
plt.legend()
plt.show()
```
